## Data exploration using visualization

### Dataset:
The dataset is obtained from FiveThirtyEight [Github repo](https://github.com/fivethirtyeight/data/tree/master/college-majors), they collected data from [American Community Survey](https://www.census.gov/programs-surveys/acs/) cleaned it and published it on their website. The dataset is on the job outcomes of students who graduated from college between 2010 and 2012. Here are some of the columns in the dataset:

 - Rank - Rank by median earnings (the dataset is ordered by this column).
 - Major_code - Major code.
 - Major - Major description.
 - Major_category - Category of major.
 - Total - Total number of people with major.
 - Sample_size - Sample size (unweighted) of full-time.
 - Men - Male graduates.
 - Women - Female graduates.
 - ShareWomen - Women as share of total.
 - Employed - Number employed.
 - Median - Median salary of full-time, year-round workers.
 - Low_wage_jobs - Number in low-wage service jobs.
 - Full_time - Number employed 35 hours or more.
 - Part_time - Number employed less than 35 hours.

### Using visualizations, I explored following questions about the dataset

 - Do students in more popular majors make more money?
    - Using scatter plots
 - How many majors are predominantly male? Predominantly female?
    - Using histograms
 - Which category of majors have the most students?
    - Using bar plots

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Checking missing data and removing it
raw_data_count = recent_grads.shape[0]
print(raw_data_count)
```

```python
recent_grads = recent_grads.dropna(axis=0)
```

I generated scatter plots in separate cells to explore the following relations:
 - Sample_size and Median
 - Sample_size and Unemployment_rate
 - Full_time and Median
 - ShareWomen and Unemployment_rate
 - Men and Median
 - Women and Median

```python
ax = recent_grads.plot(x='Sample_size', y='Employed', kind='scatter')
ax.set_title('Sample_size vs. Employed')
```

```python
ax = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax.set_title('Unemployment_rate vs. Sample_size')
```

```python
recent_grads['Sample_size'].plot(kind='hist')
```

```python
recent_grads['Sample_size'].hist(bins=25, range=(0,5000))
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,18))
for r in range(1,5):
    ax = fig.add_subplot(4,1,r)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=45, bins=20)
    ax.set_title(cols[r])
```

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(5,12))
for r in range(4,8):
    ax = fig.add_subplot(4,1,r-3)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=30)
```
